```python
import pandas as pd

df = pd.read_csv("student_performance.csv")
print(df.head())
print(df.info())
```

```python
# save_model.py
import joblib
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import DBSCAN
import pandas as pd

# Assume your dataframe is df and features are selected
features = ['StudyHours', 'Attendance', 'OnlineCourses',
            'AssignmentCompletion', 'StressLevel', 'Motivation']

X = df[features]

# Scale
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Optional: reduce dimensions
pca = PCA(n_components=3)
X_reduced = pca.fit_transform(X_scaled)

# Train DBSCAN
dbscan = DBSCAN(eps=1.5, min_samples=30)
dbscan.fit(X_reduced)

# Save the objects
joblib.dump(scaler, 'scaler.pkl')
joblib.dump(pca, 'pca.pkl')
joblib.dump(dbscan, 'dbscan.pkl')
joblib.dump(X_reduced, 'X_reduced.pkl')  # preprocessed training data for app
joblib.dump(dbscan.labels_, 'labels_train.pkl')  # labels of training points
```
